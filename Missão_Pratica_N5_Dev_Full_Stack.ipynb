## **2-Geração dos dados**

```python
import faker

arquivo = 'Nome,Pontuacao' + '\n'
fake = faker.Faker('pt-Br')

for _ in range(10):
  nome = fake.name()
  pontuacao = fake.random_int(1,10)
  arquivo = arquivo + f'{nome}' + ',' + f'{pontuacao}' +'\n'
```

## **3-Criação do arquivo de texto**

```python
%%writefile teste.txt
 
```

## **4-Gravação dos dados no arquivo txt**

```python
with open('teste.txt', 'w') as fl:
    fl.write(arquivo)
```

## **5-Recuperação do arquivo CSV e criação do histograma**

```python
import csv

txtfile = 'teste.txt'
Nomes = []
Pontuacoes = []

try:
    f = open(txtfile, 'r')
    reader = csv.DictReader(f, delimiter=',')

    for row in reader:
        Nomes.append(row['Nome'])
        Pontuacoes.append(int(row['Pontuacao']))
finally:
    f.close()
```

## **6-Manipulação dos dados para visualizá-los em um histograma**

```python
import matplotlib.pyplot as plt

n, bins, colunas = plt.hist(Pontuacoes, facecolor='blue',alpha=0.75, density=True); 

plt.xlabel('Pontuações')
plt.ylabel('Probabilidade')
plt.title('Histograma das Pontuações')
plt.grid(True)
plt.show()
```

## **7-Manipulação os dados para visualizá-los em uma nuvem de palavras**

```python
from wordcloud import WordCloud

nExtenso = {1:'um', 2:'dois', 3:'tres',4:'quatro',
            5:'cinco',6:'seis',7:'sete',8:'oito',
            9:'nove', 10:'dez'}

NewPont = []

for key, value in nExtenso.items():
    if key in Pontuacoes:
      NewPont.append(value)

nuvem_palavras = WordCloud(background_color='black',
                           width=800,height=400).generate(' '.join(NewPont))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis("off")
nuvem_palavras.to_file("Nuvem de palavras.png")
```
